# Fazendo algumas análises de indicadores com gráficos

```python
#Quantas vendas foram realizadas pra cada plataforma, considerando todos os anos?
vg_sales_by_Platform = vg_sales.Platform.value_counts()
vg_sales_by_Platform
```

```python
vg_sales_revenue_platform_year = vg_sales[['Platform','Year','Global_Sales']].groupby(['Platform','Year']).sum('Global_sales')
vg_sales_revenue_platform_year.reset_index(inplace=True)
vg_sales_revenue_platform_year
#na primeira linha, pode-se adicionar ".sort_values('Global_Sales', ascending=False) para ordenar Global_Sales em ordem decrescente"
```

```python
plt.figure(figsize=(15,10))
plt.scatter(vg_sales_revenue_platform_year.Platform,vg_sales_revenue_platform_year.Year,vg_sales_revenue_platform_year.Global_Sales)
```
